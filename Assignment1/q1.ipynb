```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

Define the math function

```python
def func(x1, x2):
    y = x1**2 + x2**2 + x1*x2
    result = (-1)*np.cos(y)
    return result
```

Contour Plot

```python
x1 = np.arange(-2,2)
x2 = np.arange(-2,2)

[X1, X2] = np.meshgrid(x1, x2) 
fig, ax = plt.subplots(1, 1)

Z = (-1)*np.cos(X1**2 + X2**2 + X1*X2)

ax.contour(X1, X2, Z) 
  
ax.set_title('Contour Plot') 
ax.set_xlabel('x1') 
ax.set_ylabel('x2') 
  
plt.show() 

# source: https://www.geeksforgeeks.org/contour-plot-using-matplotlib-python/
```

Compute gradients

```python
# Gradient 
x1, x2 = sp.symbols('x1 x2')
f = (-1)*sp.cos(x1**2 + x2**2 + x1*x2)
grad_f = [sp.diff(f, var) for var in (x1, x2)]
print(grad_f)  
```
